```python
import pandas as pd

#ImportaÃ§Ã£o das Bases
produtos_df = pd.read_csv('Contoso - Cadastro Produtos.csv', sep = ';', encoding='latin1')
vendas_df = pd.read_csv('Contoso - Vendas - 2017.csv', sep = ';', encoding='latin1')
lojas_df = pd.read_csv('Contoso - Lojas.csv', sep = ';', encoding='latin1')
clientes_df = pd.read_csv('Contoso - Clientes.csv', sep = ';', encoding='latin1')

#Renomeando Colunas
try:
    produtos_df.rename(columns={'Ã¿Nome do Produto' : 'Nome do Produto'}, inplace = True)
    lojas_df.rename(columns={'Ã¿ID Loja' : 'ID Loja'}, inplace = True)
    clientes_df.rename(columns={'Ã¿ID Cliente' : 'ID Cliente'}, inplace = True)
except:
    print("Ja foi alterado")

#Pegando apenas as colunas necessÃ¡rias
try:
    clientes_df = clientes_df[['ID Cliente', 'E-mail']]
    produtos_df = produtos_df[['ID Produto', 'Nome do Produto']]
    lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]
except:
    print("AlteraÃ§Ã£o de colunas jÃ¡ feitas")

#Juntando bases
try:
    vendas_df = vendas_df.merge(produtos_df, on = 'ID Produto')
    vendas_df = vendas_df.merge(lojas_df, on = 'ID Loja')
    vendas_df = vendas_df.merge(clientes_df, on = 'ID Cliente')
except:
    print("Merge jÃ¡ feito")


display(vendas_df)
```
